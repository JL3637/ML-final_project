```python
import numpy as np
import pandas as pd

data = pd.read_csv("train.csv")
keys = ['Energy', 'Key', 'Loudness', 'Speechiness',
       'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo',
       'Duration_ms', 'Views', 'Likes', 'Stream', 'Album_type', 'Licensed','Comments']

famous_keys = ['Views', 'Likes', 'Stream', 'Comments']
for key in keys:
    if key == 'Album_type' or key == 'Licensed':
        continue
    if key in famous_keys:
        data[key].fillna(value=0,inplace=True)
    else:
        data[key].fillna(value=data[key].mean(),inplace=True)
y = data['Danceability']
new_data = {}
for key in keys:
    if key != 'Danceability':
        new_data[key] = data[key]
data = pd.DataFrame(new_data)
data
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import normalize
def datatoX(data):
    famous_keys = ['Views', 'Likes', 'Stream', 'Comments']
    famous_X = []
    temp_X = []
    type_X = []
    for key in data:
        if key == 'Album_type':
            types = data['Album_type'].unique()
            for t in types:
                type_X.append(np.array([1 if ty==t else 0 for ty in data['Album_type']]))
        elif key == 'Licensed':
            type_X.append(np.array([1 if l else 0 for l in data['Licensed']]))
        elif key in famous_keys:
            famous_X.append(np.array([np.log(max(a,0.000001)) for a in data[key]]))       
        elif key in keys:
            temp_X.append(np.array([a for a in data[key]]))       
#             temp_X.append(np.array([1/(a+0.000001) for a in data[key]]))
    temp_X = normalize(temp_X)    
    famous_X = normalize(famous_X)
    type_X = np.array(type_X)
    poly = PolynomialFeatures(2)
    temp_X = poly.fit_transform(temp_X.T)
    famous_X = np.array(famous_X)
    X = np.concatenate((temp_X,famous_X.T,type_X.T),axis=1)
    print(X.shape)
    return X
    
X = datatoX(data)
X
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error

def test1(X, y, X_test):
    kf = KFold(n_splits=10, random_state=42, shuffle=True)
    predictions_array = []
    CV_score_array    = []
    for train_index, valid_index in kf.split(X):
        X_train, X_valid = X[train_index], X[valid_index]
        y_train, y_valid = y[train_index], y[valid_index]
        regressor = HistGradientBoostingRegressor(loss='absolute_error', learning_rate=0.1, max_iter=400, max_leaf_nodes=200)
        regressor.fit(X_train, y_train)
        predictions_array.append(regressor.predict(X_test))
        CV_score_array.append(mean_absolute_error(y_valid,regressor.predict(X_valid)))
    predictions = np.mean(predictions_array,axis=0)
    predictions[predictions<0] = 0
    predictions[predictions>9] = 9
    print(f"CV score {np.min(CV_score_array)}")
    return np.rint(predictions)
```
